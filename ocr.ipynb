```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from PIL import Image
from pytesseract import *
%matplotlib inline

image_file = 'image.jpg'
image_file = Image.open(image_file)

width, height = image_file.size

plt.rcParams['figure.figsize'] = [200, 100]

im = np.array(image_file) #Image.open(image_file, dtype=np.uint8))

# Create figure and axes
fig, ax = plt.subplots(1)

# Display the image
ax.imshow(im)

boxes = image_to_boxes(image_file)
boxes = boxes.split("\n")
#print(len(boxes))
for box in boxes :
    box = box.split(" ")

    # Create a Rectangle patch
    rect = patches.Rectangle((int(box[1]),height-int(box[2])),int(box[3])-int(box[1]),-(int(box[4])-int(box[2])),linewidth=1,edgecolor='r',facecolor='none')

    # Add the patch to the Axes
    ax.add_patch(rect)
    
#plt.show()
plt.savefig('image_boxes.jpg')
```
